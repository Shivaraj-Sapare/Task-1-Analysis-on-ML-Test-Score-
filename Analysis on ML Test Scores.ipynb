```python
import pandas as pd
file_path = "/content/scores_data.csv"
df = pd.read_csv(file_path)
df.columns = df.columns.str.strip()
df['Score'] = df['Score'].apply(lambda x: int(x.split('/')[0].strip()))
print(df.head(), "\n")
grouped = df.groupby("Batch")["Score"]
summary = grouped.agg(
    Students="count",
    Average="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Std_Deviation="std"
)
summary = summary.round(2)
print("Batch-wise Performance Summary:\n")
print(summary)
ranking = summary.sort_values(by="Average", ascending=False)
print("\n Batch Ranking (Best to Weakest):\n")
print(ranking[["Average"]])
print("\n Key Insights:")
best_batch = ranking.index[0]
weak_batch = ranking.index[-1]
print(f"- Best performing batch: {best_batch}")
print(f"- Weakest performing batch: {weak_batch}")
high_variability = summary["Std_Deviation"].idxmax()
print(f"- Batch with highest score variability: {high_variability}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("scores_data.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df['score'] = df['score'].apply(lambda x: int(x.split('/')[0].strip()))
grouped = df.groupby("batch")["score"]
avg_scores = grouped.mean()
plt.figure()
avg_scores.plot(kind="bar")
plt.title("Average Score by Batch")
plt.xlabel("Batch")
plt.ylabel("Average Score")
plt.show()
plt.figure()
df.boxplot(column="score", by="batch")
plt.title("Score Distribution by Batch")
plt.suptitle("")
plt.xlabel("Batch")
plt.ylabel("Score")
plt.show()
```
